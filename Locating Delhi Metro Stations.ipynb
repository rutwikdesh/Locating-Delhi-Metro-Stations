# Delhi Metro Stations

### Import Relevant Libraries

```python
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('DELHI_METRO_DATA.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df['Line'].value_counts()
```

```python
latitude, longitude = df['Latitude'].median(), df['Longitude'].median()
```

### Using Folium to Visualize the Delhi Metro stations geographically

```python
import folium
```

```python
map1 = folium.Map(location=[latitude, longitude])

for lat, long, label in zip(df['Latitude'], df['Longitude'], df['Station']):
    folium.features.CircleMarker(
        [lat,long],
        radius = 5,
        color = 'blue',
        popup = label,
        fill=True,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(map1)
map1
```
